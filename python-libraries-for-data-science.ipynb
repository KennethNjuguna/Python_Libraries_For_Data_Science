# 1. NUMPY

**Numpy** is the core library for scientific and Numerical computing in Python. It provides high performance multidimesnsional array object and tools for working with arrays. 

* Numpy main object is the Multidimensional array.
* It is a table of elements (usually numbers), all of the same type, indexed by a tuple of intergers.
* In numpy dimensions are called *Axes*

In numpy we can we have a one dimension array and two dimensional arrays.
This arrays come with different shapes.
For example a one dimensional array will come with a shape of shape: (6,).
While a two dimensional array will have a shape: (2,4).






Numpy is better than Python list because it is **Fast, Convenient and Less memory**

Comparison between a Python List and Numpy is as shown below. First, we import time and system to check at how both execute in regards to time taken and Memeory used.

```python
import time
import sys
```

We can check the memory as below

```python
#Create a list
b=range(1000)
print(sys.getsizeof(5)*len(b))
```

```python
#Create a numpy
c=np.arange(1000)
print(c.size*c.itemsize)
```

Above as you can observe the Python list is consuming more memory as compared to the numpy.
Below we can check the speed and compare a python list vs a numpy.

```python
size=100000
```

```python
L1=range(size)
L2=range(size)
A1=np.arange(size)
A2=np.arange(size)
```

Lets check the time it takes to execute a Python list.

```python
start=time.time()
result=[(x+y)for x,y in zip(L1,L2)]
result=[(x+y)for x,y in zip(L1,L2)]
print(result)
print("python list took:",(time.time()-start)*1000)
```

Lets check the time it takes to execute numpy array.

```python
start=time.time()
result= A1+A2
print("Numpy array took:",(time.time()-start)*1000)
```

From the above numpy takes quick and it is more fast as compared to Python list.

**Draw a contour image**

```python
import matplotlib
import numpy as np
import matplotlib.cm as cm
import matplotlib.pyplot as plt

delta=0.25
x = np.arange(-3.0,3.0,delta)
y = np.arange(-2.0,2.0,delta)
X, Y = np.meshgrid(x,y)
Z1= np.exp(-X**2-Y**2)
Z2= np.exp(-(X-1)**2-(Y-1)**2)
Z = (Z1-Z2)*2

print(X)
print(Y)

#Plot the contour
fig, ax = plt.subplots()
CS = ax.contour(X,Y,Z)
ax.clabel(CS, inline=1, fontsize=10)
ax.set_title('Contour Map')
```

* Simple and efficient tool for data mining and data analysis.
* Built on NumPy, SciPy and Matplotlib.
* Open source, commercially stable BSD- Burkley  license.

**What we achieve using scikit-learn**
* **Classification** Identifying which category an object belongs to. Application: Spam Detection.
* **Regression** Predicting an attribute associated with an object. Application: Stock Prices prediction
* **Clustering** Automatic grouping of similar objects to sets. Application: Customer segmentation.
* **Model selection** Comparing, Validating and choosing parameters and models. Application: Improving accuracy via parameter tuning.
* **Dimensionality Reduction** Reducing the number of random variables to consider. Application: To increase model efficiency.
* **Pre-Processing** Feature extraction and normalization. Application: Transforming input data such as text for use with machine learning extractions.
